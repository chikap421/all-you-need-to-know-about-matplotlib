## 1.1 Methods of plotting using matplotlib.pyplot

```python
# The first method of plotting
# A longer method of doing plt.plot()
fig = plt.figure () #creates a figure
ax = fig.add_subplot() #add some axis to the figure
ax.plot(x,y) # pass some data into the plot
plt.show()
```

```python
# The 2nd method of plotting
fig = plt.figure() # create a figure
ax = fig.add_axes([1,1,0.5,0.4]) # add some axes with arguments that define the left, bottom, width and height dimensions
ax.plot(x,y)
plt.show()
```

```python
## The 3rd method of plotting (recommedned method) - Object oriented method of plotting.
fig, ax = plt.subplots()
ax.plot(x, [50, 129, 400, 1000, 2e3]); #add some data
#The plot autmomatically adjusts to the data entries

# To undertstand why this method is called the object-oriented programming method
type(fig), type(ax)

# The Figure object and the AxesSubplot objects are included. That is why it is called object-oriented plotting
```

## 1.2 Matplotlib typical workflow

```python
# 1. import matplotlib library and make it available for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Get the data ready for plotting
x = [1, 2, 5, 8, 10]
y = [100, 200, 500, 600, 1000]

#3. Setup plot using the object-oriented method
fig, ax = plt.subplots(figsize = (5, 5)) # Set the width, height of the plot

# 4. Plot the data
ax.plot (x, y) # Add some data to the axis you have created

#5. Customize the plot by provding relevant information
ax.set(title = 'Simple plot',
      xlabel = 'x-axis',
      ylabel = 'y-axis')

#6. Save the figure and show it
fig.savefig('figures/sample-plot.png', dpi=300) # dpi sets the image quality
```

## 3. Two ways of creating multiple plots in one plot


```python
# Option 1 for subplots (The tuples way)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10,5))

# To put data on the different plots
ax1.plot(x, x/2);
ax2.scatter(np.random.rand(10), np.random.randn(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Option 2 for subplots (the least index way)
fig, ax = plt.subplots(nrows = 2,
                     ncols = 2,
                     figsize = (10, 5))
# Plot into each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter (np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# 5. Choosing between the object oriented (OO) method and the pyplot (.plot) method

* The pyplot method is better to use when plotting something quickly
* The OO method is better to use when making advanced plots

```python
heart_disease.head()
```

```python
# To see the patients over 50 using boolean indexing
over_50=heart_disease[heart_disease['age'] > 50]
over_50.head()
```

## 5.1 Pyplot (.plot) method AKA plotting directly from pandsas DF

```python
# Make a scatter plot of age and cholestrol
# Pyplot (.plot) method AKA plotting directly from pandas DF
over_50.plot(kind='scatter',
            x ='age',
            y='chol',
            c='target');
# c = 'target' means to use the target column as the colorbar

#Further reading: https://stackoverflow.com/questions/52108558/how-does-parameters-c-and-cmap-behave-in-a-matplotlib-scatter-plot
```

## 5.2 Object oriented (OO) method + pylpot method

```python
# Using the OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);

# ax = ax is used to assign the plot to the current axes

# to change the limit of the x-axes from 45-100
ax.set_xlim([45, 100]);
```

## 5.3 Pure Object oriented (OO) method

```python
#To view the target column entries
over_50['target'].values
```

 More ways of adding `legends` to plots can be found here
 https://stackoverflow.com/questions/19125722/adding-a-legend-to-pyplot-in-matplotlib-in-the-simplest-manner-possible

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);
# c=target means let the marker be determined by the target column

# Customize the plots
ax.set(title='Display of the heart disease and cholestrol levels',
      xlabel='Age',
      ylabel='Cholestrol');

# To add a legend for the target column
ax.legend(*scatter.legend_elements(), title='Target');

# To add a horizontal mean line 
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# To add subplopts to the existing plot (chol and thalach vs age)


# Create a single figure with two axes
# Use sharex=True for both plots to have the same x axes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True);

# Add the chol data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target']);

# Customize ax0
ax0.set(title='Cholestrol and heart disease levels',
       ylabel='Cholestrol');
# Add a legend
ax0.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal mean line
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');

# Add the thalach data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target']);

# Customize ax1
ax1.set(title='Maximum heart rate and heart disease levels',
        xlabel = 'Age',
        ylabel = 'Maximum heart rate');
# Add a legend
ax1.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal mean line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# To add a title to the entire figure
fig.suptitle('Heart disease visualization', fontsize=16, fontweight='bold');
```

### 6.3 Changing a style within a style and using different color

You can change the **color** in your plots using the `cmap` function.
More about the documentation can be found here:
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
# Set the plot style
plt.style.use('seaborn-whitegrid')

# To add subplopts to the existing plot (chol and thalach vs age)


# Create a single figure with two axes
# Use sharex=True for both plots to have the same x axes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True);

# Add the chol data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                     cmap='winter'); # Change the color scheme of the plot

# Customize ax0
ax0.set(title='Cholestrol and heart disease levels',
       ylabel='Cholestrol');
# Add a legend
ax0.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal mean line
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');

# Add the thalach data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                     cmap='autumn'); # Change the color

# Customize ax1
ax1.set(title='Maximum heart rate and heart disease levels',
        xlabel = 'Age',
        ylabel = 'Maximum heart rate');
# Add a legend
ax1.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal mean line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# To add a title to the entire figure
fig.suptitle('Heart disease visualization', fontsize=16, fontweight='bold');
```

### 6.4 Customizing the x and y axis limitations in a plot

```python
# To add subplopts to the existing plot (chol and thalach vs age)
# Create a single figure with two axes
# Use sharex=True for both plots to have the same x axes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True);
# Add the chol data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                     cmap='winter');
# Customize ax0
ax0.set(title='Cholestrol and heart disease levels',
       ylabel='Cholestrol');
#Change the ax0 x-axis limits
ax0.set_xlim([50,80])
# Add a legend
ax0.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal mean line
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');


# Add the thalach data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                     cmap='winter');
#Change the ax1 y-axis limits
ax1.set_ylim([60,200])
# Customize ax1
ax1.set(title='Maximum heart rate and heart disease levels',
        xlabel = 'Age',
        ylabel = 'Maximum heart rate');
# Add a legend
ax1.legend(*scatter.legend_elements(), title='Target');
# Add a horizontal mean line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# To add a title to the entire figure
fig.suptitle('Heart disease visualization', fontsize=16, fontweight='bold');
```
